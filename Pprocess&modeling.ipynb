```python
import json
with open('symptoms.json') as f:
    symptoms = json.load(f)
symptoms = [symptoms]
X = data.iloc[:,4].values

for i in range(len(X)):
    symptoms.append([])
    for j in range(len(symptoms[0])):
        if symptoms[0][j] == X[i]:
            symptoms[i+1].insert(j, 1)
        else:
            symptoms[i+1].insert(j, 0)
            
for k in range(5,29):
    X = data.iloc[:,k].values

    for i in range(len(X)):
        for j in range(len(symptoms[0])):
            if symptoms[0][j] == X[i]:
                symptoms[i+1][j] = 1
                
header = symptoms.pop(0)
df = pd.DataFrame(symptoms, columns=header)

df.replace(0, pd.np.nan, inplace=True)
df.dropna(axis=1, how='all', inplace=True)
df.replace(df.iloc[0,0], 0, inplace =True)
df = df.loc[:,~df.columns.duplicated()]
df = df.drop_duplicates(keep='first')
df
```

```python
data1 = data[['long_title','age','gender','bmi']]
data1.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)
df_final = pd.concat([data1,df], axis=1)
df_final.dropna(inplace=True)
df_final
```

```python
from  sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

X = df_final.iloc[:,1:].values
Y = df_final.iloc[:,0].values

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.033)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test =  sc.transform(X_test)

clf = RandomForestClassifier(n_estimators=10)
clf.fit(X, Y)
y_pred = clf.predict(X_test)
print('Score=',clf.score(X=X_test, y=y_test))

a = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print('accuracy=',a*100,'%')
```
