```python
import re
import pandas as pd
import psycopg2
sqluser = 'postgres'
sqlpass = 'postgres'
dbname = 'mimic'
schema_name ='mimiciii'
host = 'localhost'
query_schema = 'SET search_path to ' + schema_name + ';'

con = psycopg2.connect(dbname=dbname, user=sqluser, password=sqlpass, host=host)
query = query_schema + """
select text from noteevents where hadm_id = '167853' and category = 'Discharge summary' 
"""
df = pd.read_sql_query(query,con)
df
```

```python
 pd.options.display.max_colwidth = 10000000
```

```python
df2 = df.replace(to_replace =r'\n', value = ' ', regex = True)
```

```python
text = df2.to_string(header=False,
                  index=False,
                  index_names=False)
text
```

```python
preprocessing = re.findall("confirmed[^by|with] (.*?)\.",text)#This denotes the symptoms confirmed by the medical officials
preprocessing2 = re.findall("showed (.*?)\.",text)#This denotes the symptoms confirmed during the patient's treatment
preprocessing3 = re.findall("reports (.*?)\.",text)#This denotes any symptoms that the first two regex sweeps may have missed
preprocessing.extend(preprocessing2)
preprocessing.extend(preprocessing3)
newpreprocess = []

for i in preprocessing:
    if ',' in i:
        newpreprocess.extend(i.split(", "))
        for j in i.split(","):
            if re.findall(". .*",j):
                newpreprocess.extend(re.findall("[a-z] (.*)",j))#Attempting to clear a typing mistake from ruining the data, failed, will probably go away with NLP usage 
            else:
                newpreprocess.append(j)
    else:
        newpreprocess.append(i)
print(newpreprocess)
```
